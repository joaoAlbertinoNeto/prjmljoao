```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder

# Carregue o dataset
df = pd.read_csv("2015-acessos.csv")

# Crie uma cópia só com as colunas relevantes
df_model = df[["UF", "Tecnologia Geração", "Modalidade de Cobrança", "Tipo de Pessoa", "Acessos"]].copy()

# Codificar colunas categóricas
le = LabelEncoder()
for col in ["UF", "Tecnologia Geração", "Modalidade de Cobrança", "Tipo de Pessoa"]:
    df_model[col] = le.fit_transform(df_model[col].astype(str))

# Treinar Isolation Forest
model = IsolationForest(contamination=0.01, random_state=42)
df_model["anomaly"] = model.fit_predict(df_model)

# Juntar a coluna de anomalias com o dataset original
df["anomaly"] = df_model["anomaly"]

# Ver municípios considerados anômalos
anomalias = df[df["anomaly"] == -1]
print(anomalias[["Município", "UF", "Acessos", "Tecnologia Geração", "Modalidade de Cobrança"]])
```
